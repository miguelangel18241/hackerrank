```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Define the hexadecimal color extracted from the image
average_hex_color = '#3e4757'  # RGB (62, 71, 87)

# Define the text for the word cloud
text = """
Python Python Python Python Python Data Analysis JavaScript HTML CSS SpringBoot
Communication Team Collaboration Attention to Detail Problem-Solving Computer Literacy
ASX Finance MachineLearning AI Software Development Cloud Computing
"""

# Custom color function to use only white and the extracted color from the image
def custom_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return average_hex_color if random_state.randint(0, 2) else "white"

# Generate the word cloud using the custom color function
wordcloud = WordCloud(width=800, height=400, background_color='white', max_font_size=100, max_words=50,
                      color_func=custom_color_func).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
